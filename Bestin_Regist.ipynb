```python
#-*-coding: utf-8-*-

#-*-coding: euc-kr-*-
import requests
AUTH_URL = "https://center.hdc-smart.com"
CONTROL_URL = "control url"
UUID = "123123" # Unique한 ID로 변경 필요.
headers = {"Content-Type":"application/json", "Authorization":UUID}

def printResponse(response):
    print(response)
    print(response.content)

def registration():
    datas = {
        "site" : "Site Code", #AUTH_URL에서 본인 아파트 찾은 후, 자세히 버튼 누르면 나오는 코드
        "identifier":"", # 본인 동/호수 ex.)501/1301
        "alias":"HA" # 월패드에 보일 Label
    }

    url = AUTH_URL + "/v3/auth/registration"
    response = requests.post(url, headers=headers, json=datas)
    printResponse(response)
registration()
```

```python
def verify():
    datas = {
        "transaction" : "", # registration에서 출력된 코드
        "password" : "" # 월패드에 표시된 인증번호
    }
    url = AUTH_URL + "/v3/auth/verify"
    response = requests.post(url, headers=headers, json=datas)
    printResponse(response)
```

```python
verify()
```
